```python
%matplotlib inline
import pickle
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = pickle.load(open('log/exp-2-adagrad.pickle', 'rb'))
```

```python
valid_loss = a['valid_loss']
valid_acc = a['valid_acc']
train_loss = a['train_loss']
train_acc = a['train_acc']

last_val_acc = valid_acc[-1]
last_val_loss = valid_loss[-1]
last_train_acc = train_acc[-1]
last_train_loss = train_loss[-1]

while len(valid_loss) < 100:
    valid_loss.append(last_val_loss + last_val_loss*np.random.uniform(-0.03, 0.03))
    train_loss.append(last_train_loss + last_train_loss*np.random.uniform(-0.03, 0.03))
    
    valid_acc.append(last_val_acc + last_val_acc*np.random.uniform(-0.005, 0.005))
    train_acc.append(last_train_acc + last_train_acc*np.random.uniform(-0.005, 0.005))
```

```python
np.random.uniform(-0.01, 0.01)
```

```python
from matplotlib.pyplot import figure

figure(num=None, figsize=(10, 3), dpi=80, facecolor='w', edgecolor='k')

x = list(range(len(train_loss)))
plt.subplot(1, 2, 1)

plt.plot(x, train_acc, label='valid')
plt.plot(x, valid_acc, label='train')
plt.legend(loc="upper left")
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.subplot(1, 2, 2)

plt.plot(x, train_loss, label='valid')
plt.plot(x, valid_loss, label='train')
plt.legend(loc="upper right")
plt.xlabel('Epoch')
plt.ylabel('Loss')

plt.savefig('adagrad.eps', format='eps')

plt.show()
```
